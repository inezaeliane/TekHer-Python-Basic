### Exercise 40:
Create a stacked area plot using three random datasets.


```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values (e.g., time or index)
x = np.arange(0, 10, 1)

# Generate three random datasets
np.random.seed(0)  # For reproducibility
y1 = np.random.randint(1, 10, size=len(x))
y2 = np.random.randint(1, 10, size=len(x))
y3 = np.random.randint(1, 10, size=len(x))

# Stack the datasets into one array
y = np.vstack([y1, y2, y3])

# Create the stacked area plot
plt.stackplot(x, y1, y2, y3, labels=['Dataset 1', 'Dataset 2', 'Dataset 3'], alpha=0.8)
plt.legend(loc='upper left')
plt.title('Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
import matplotlib.pyplot as plt

# 1. Data
transactions = ['Start', 'Sales', 'Service', 'Costs', 'End']
amounts = [100, 50, 20, -40, 130]
is_total = [True, False, False, False, True]

# 2. Process data to find bar bottoms
bottoms = []
running_total = 0
for amount, is_total_flag in zip(amounts, is_total):
    if is_total_flag:
        bottoms.append(0)
        running_total = amount
    else:
        bottoms.append(running_total)
        running_total += amount

# 3. Plotting
fig, ax = plt.subplots()
ax.set_title('Simple Waterfall Chart')

# Define colors based on value
colors = ['b' if total else ('g' if amt > 0 else 'r') 
          for amt, total in zip(amounts, is_total)]

# Plot the bars
ax.bar(transactions, amounts, bottom=bottoms, color=colors)

# Add connecting lines
for i in range(1, len(amounts)):
    if not is_total[i]:
        line_y = bottoms[i]
        # Connect previous bar to current bar
        ax.plot([i - 0.5, i + 0.5], [line_y, line_y], color='gray', linestyle='--')

ax.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
